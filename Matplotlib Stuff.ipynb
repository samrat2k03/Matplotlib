```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("50_Startups.csv")
print(df)
```

```python
df.info()
```

```python
#  it shows the no. of null value in csv file
df.isnull().sum()
```

```python
state = df['State'].unique()
state
```

```python
profit = df['Profit'].max()
print(profit)
idx = df['Profit'].idxmax()
profit = df.loc[idx, 'State']
print(profit)
```

```python
x = df['Profit']
y = df['R&D Spend']

### line plot
show_fig = plt.plot(x,y)
plt.title("Profit VS R&D Spend")
plt.xlabel("Profit")
plt.ylabel("Spends")
plt.show()
```

```python
# ### bar Plot

plt.bar([0.25,2.25,3.25,5.25,7.25],[300,400,200,600,700],
label="Carpenter",color='b',width=0.5)
plt.bar([0.75,1.75,2.75,3.75,4.75],[50,30,20,50,60],
label="Plumber", color='g',width=.5)
plt.legend()
plt.xlabel('Days')
plt.ylabel('Wage')
plt.title('Details')
```

```python
### Scatter plot
plt.scatter(x,y, color = 'red')
```

```python
### PieChart

RD_Spend1 = df['R&D Spend'].max()
Administration = df['Administration'].max()
Marketing_Spend = df['Marketing Spend'].max()
Profit1 = df['Profit'].max()

parts = [RD_Spend1,Administration,Marketing_Spend,Profit1]
labels = ['R&D Spend','Administration','Marketing Spend','Profit']
plt.title("Maximum Values")
plt.pie(parts, labels=labels,autopct ='%1.1f%%')
```
